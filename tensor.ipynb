#creating Tensor with tf.variable()
the difference between tf.variable() and tf.constant(),is tf.constant() are immutable (can't be changes,only be used to create new tensor),where as tensor created with tf.variable() are mutable.


```python
#creating a constant() and variable() with same
changeable_tensor_var=tf.Variable([10,6])
unchangeable_tensor_var=tf.constant([10,6])
changeable_tensor_var,unchangeable_tensor_var
```

```python
#using try and except method to handle the error.To change the value we need to use assign() method
#variable() is mutable as the value is changed
try:
  changeable_tensor_var[0].assign(3)
except Exception as e:
  print(e)
changeable_tensor_var
```

```python
#using try and except method to handle the error.We're trying to change the constant()
#which cause the error as it's immutable
try:
  unchangeable_tensor_var[1].assign(3)
except Exception as e:
  print(e)
```

#Creating random tensors
random tensors are tensors of some abitrary size which contain random numbers.

purpose of random tensors is that neural networks use to intialize their weights(patterns)that they're trying to learn in the data.

The process of neural network learning often involves taking a random n-d array of numbers and refining them unitl they represent some kind of pattern

```python
#creating random tensors
va=tf.random.Generator.from_seed(42)
va1=va.normal(shape=(2,1))
va1
```

```python
#creating a 2 random(but the different)tensors
va2=tf.random.Generator.from_seed(42)
va2=va2.normal(shape=(2,1))
va3=tf.random.Generator.from_seed(1)
va3=va3.normal(shape=(2,1))
va2,va3
```

```python
#creating a 2 random(but the same)tensors
va2=tf.random.Generator.from_seed(42)#set the seed for reproduciability
va2=va2.normal(shape=(2,1))#creating tensors from normal distrubtion
va3=tf.random.Generator.from_seed(42)
va3=va3.normal(shape=(2,3))
va2,va3,va2==va3
```

```python
#shuffling the tensors using shuffle()
s=tf.random.shuffle([[1,2],
                     [3,5],[5,5]])
s
```

```python
#setting the shuffel order same by using seed()
tf.random.set_seed(42)#global random seed()
s1=tf.random.shuffle([[1,2],
                     [3,5],[6,5]],seed=42)#set the seed as global random see()
s1
```

#Matrix Multiplication
TensorFLow implemetns matrix multiplication functionality in the tf.matmul() method and other way  with python opertor "@".

the main 2 rules in matrix multiplications is:

1.The number of columns in the first matrix must be equal to the number of rows in the second matrix

2.The resulting matrix, known as the matrix product, has the number of rows of the first and the number of columns of the second matrix.

```python
a=tf.random.Generator.from_seed(42)
a=a.normal(shape=(3,3))
a
```

```python
#using tensorflow built-in funtion matmul() to multiply matrix
tf.matmul(a,a)
```

```python
#using python operator "@" for matrix multiplcation
a@a
```

#reshaping the matrix ,so it match the inner dimensions
 ex:if you have a=(2,3) and b=(5,3) shape of matix,you need to reshape the matrix b to become the b=(3,5) to match the inner dimensions(column of 1st matix and row of 2nd matrix)


```python

a=tf.random.Generator.from_seed(42)
a=a.normal(shape=(2,3))
b=tf.random.Generator.from_seed(42)
b=b.normal(shape=(5,3))
a,b
```

```python
#using reshape()-allows in tensor to reshape into defined shape
b=tf.reshape(b,shape=(3,5))
b,a
```

```python
#now we can do matmul() on a and b matrix
c=tf.matmul(a,b)
c
```

```python
#using transpose() method to match inner dimensions
d=tf.transpose(c)

d,tf.matmul(d,a)
```
